```python
from music21 import *
def play(stream):
    midi.realtime.StreamPlayer(stream).play()
```

## Nardis

```python
nardis = stream.Stream()
nardis.append(meter.TimeSignature('4/4'))
#Standard voicings here too
Em7 = chord.Chord('E G B D', quarterLength = 4)
FM7 = chord.Chord('F A C E', quarterLength = 4)
B7 = chord.Chord('B D# F# A', quarterLength = 4)
CM7 = chord.Chord('C E G B', quarterLength = 4)
Am7 = chord.Chord('A C E G', quarterLength = 4)
Dm7 = chord.Chord('D F A C', quarterLength = 4)
G7 = chord.Chord('G B D F', quarterLength = 4)
EM7= chord.Chord('E G# B D#', quarterLength = 4)

nardisA = [Em7, FM7, B7, CM7, Am7, FM7, EM7, Em7]
nardisB = [Am7, FM7, Am7, FM7, Dm7, G7, CM7, FM7]
nardisTot = nardisA + nardisA + nardisB + nardisA


#This snippet is simply for demonstrational purposes. The real work will use {@var NardisA} and {@var NardisB}.
for i in range(0, 32):
    m=stream.Measure(i)
    m.append(nardisTot[i])
    nardis.append(m)
mm1 = tempo.MetronomeMark(number=120)
nardis.append(mm1)
nardis.show()
nardis.insert(0, instrument.Piano())
play(nardis)
```

## Autumn Leaves

```python
leaves = stream.Stream()
leaves.append(meter.TimeSignature('4/4'))
Am7 = chord.Chord('A C E G', quarterLength= 4)
Em7 = chord.Chord('E G B D', quarterLength = 4)
D7 = chord.Chord('C D F# A', quarterLength = 4)
GM7 = chord.Chord('G B D F#', quarterLength = 4)
CM7 = chord.Chord('C E G B', quarterLength = 4)
Gbm7b5 = chord.Chord('F# A C E', quarterLength = 4)#Couldnt name it F#m7b5 bcs of variable naming rules, so I had to resort to Gb
B7 = chord.Chord('B D# F# A', quarterLength = 4)


#Extra chords for the C CSP, some need to be redefined because shorter length
Em72 = chord.Chord('E G B D', quarterLength = 2)
Ebm7 = chord.Chord('D# F# A# C#', quarterLength = 2)
Dm7 = chord.Chord('C D F A', quarterLength = 2)
Dbm7= chord.Chord("C# E G# B", quarterLength = 2)


leavesA = [Am7, D7, GM7, CM7, Gbm7b5, B7, Em7, Em7]
leavesB = [Gbm7b5, B7, Em7, Em7, Am7, D7, GM7, CM7, #B
           Gbm7b5, B7, Em72, Dm7, #C
           Gbm7b5, B7, Em7, Em7]

leavesTot=leavesA + leavesA + leavesB
for i in range(0, 26):
    m=stream.Measure(i)
    m.append(leavesTot[i])
    leaves.append(m)

m26 = stream.Measure(26)
m26.append([Em72, Ebm7])
m27=stream.Measure(27)
m27.append([Dm7, Dbm7])
leaves.append([m26, m27])
for i in range(28, 32):
    m=stream.Measure(i)
    m.append(leavesTot[i])
    leaves.append(m)

leaves.show()
play(leaves)

#We're choosing the standard voicing for now, we will change it when assembling the piece
```
